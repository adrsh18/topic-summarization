```python

# coding: utf-8

# In[ ]:

import tweepy as tp
import json

consumerKey = ''
consumerSecret = ''

auth = tp.OAuthHandler(consumer_key=consumerKey, consumer_secret=consumerSecret)

api = tp.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

data = []

print "Will get digging now .."

for tweet in tp.Cursor(api.search, q="%23christmas", since="2016-11-01", until="2016-12-02", count=100, lang="en").items():
    if "retweeted_status" in tweet._json:
        continue
    data.append(tweet._json)
    if len(data)%1000 == 0:
        print "Iteration done. Got ", len(data), " tweets so far!"
    if len(data) > 100000:
        print "Breaking the loop"
        break

outfile = "christmas"

print ""
print "Writing to file.."
datafile = open("%s-en.txt" % outfile,"w+")
datafile.write(json.dumps(data))
datafile.close()
print "Done : ", len(data)
```
